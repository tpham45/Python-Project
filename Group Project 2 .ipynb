```python
#SET UP FOR CURRENCY FORMATTING
import locale
locale.setlocale( locale.LC_ALL, "")

#SET NEEDED INDEXES FOR OUR FILE region, segment, category, sales, and profit
REGION_INDEX = 12
SEGMENT_INDEX = 7
CATEGORY_INDEX = 15
SALES_INDEX = 18
PROFIT_INDEX = 21
#DEFINE A CLASS
# initialize all the data in the object we are creating 
class Order:
    # Define the __init__ in order to create our object for each of our indexes 
    def __init__(self,region,segment,category,sales,profit):
        self.region = region
        self.segment = segment
        self.category = category
        self.sales = sales
        self.profit = profit
    #END INIT FUNCTION
#END CLASS ORDER

# We need to create a function that will allow us to open our file which we did just that below. 
def OpenFile():
    NAMEOFFILE = 'StoreOrders-Proj2.csv'
    #read the header
    fileIn = open(NAMEOFFILE)
    fileIn.readline()
    return fileIn
#end function OpenFile

# Create a function to process each row of data and return the columns needed for our calculation
def ReadLine(line):
    # Iterate through the each row, take out blank space with strip() method
    line = line.strip()
    # Split each column to form a list of columns
    cells = line.split(",")
    # Create new variables to call each of our indexes 
    curRegion = cells[REGION_INDEX]
    curSegment = cells[SEGMENT_INDEX]
    curCategory = cells[CATEGORY_INDEX]
    curSales = float(cells[SALES_INDEX])
    curProfit = float(cells[PROFIT_INDEX])
    curOrder = Order(curRegion, curSegment, curCategory, curSales, curProfit)
    return curOrder
# End ReadLine Function 

# Create a function to store our data from our file in list called regionList, segmentList, categoryList 
def BuildLists(dataOrders):
    # Create our list to store our data
    regionList = []
    segmentList = []
    categoryList = []
    # Create a for loop to make sure that all the data from the file is inside of our list if not then we need to 
    # append it to our list.
    for curOrder in dataOrders:
        if not curOrder.region in regionList:
            regionList.append(curOrder.region)
        # end if
        if not curOrder.segment in segmentList:
            segmentList.append(curOrder.segment)
        # end if
        if not curOrder.category in categoryList:
            categoryList.append(curOrder.category)
        # end if
    #end for loop
    return regionList, segmentList, categoryList
#End BuildLists 
        
# Created function to calculate the total region sales from our data
def CalcTotalRegionSales(regList, orderList): 
    regSales=[]
    for region in regList:
        regSales.append(0)
        curIndex = len(regSales)-1
        for order in orderList:

            if (order.region == region):
                regSales[curIndex] += order.sales
            #end if
        #end for order
    #end for region
    return regSales
#end CalcTotalRegionSales function

# Created function to calculate the total region profit from our data
def CalcTotalRegionProfit(regList, orderList):
    regProfit = []
    for region in regList:
        regProfit.append(0)
        curIndex = len(regProfit)-1
        for order in orderList:
            if (order.region == region):
                regProfit [curIndex] += order.profit
            #end if
        #end for order
    #end for region
    return regProfit
# end CalcTotalRegionProfit function

# Created function to calculate the total category sales from our data
def CalcTotalCategorySales(catList, orderList):
    catSales = []
    for category in catList:
        catSales.append(0)
        curIndex = len(catSales) - 1
        for order in orderList:
            if (order.category == category):
                catSales [curIndex] += order.sales
            #end if
        #end for order
    #end for category
    return catSales
# end CalcTotalCategorySales

# Created function to calculate the total category profit from our data
def CalcTotalCategoryProfit(catList, orderList):
    catProfit = []
    for category in catList:
        catProfit.append(0)
        curIndex = len(catProfit) -1
        for order in orderList:
            if (order.category == category):
                catProfit [curIndex] += order.profit
            #end if
        #end for order
    #end for category
    return catProfit
# end CalcTotalCategoryProfit

# Created function to calculate the total segment sales from our data 
def CalcTotalSegmentSales(segList, orderList):
    segSales =[]
    for segment in segList:
        segSales.append(0)
        curIndex = len(segSales) - 1
        for order in orderList:
            if (order.segment == segment):
                segSales [curIndex] += order.sales
            #end if
        #end for order
    #end for segment
    return segSales
# end CalcTotalSegment Sales

# Created function to calculate the total segment profit from our data
def CalcTotalSegmentProfit(segList, orderList):
    segProfit = []
    for segment in segList:
        segProfit.append (0)
        curIndex = len(segProfit) - 1
        for order in orderList:
            if (order.segment == segment):
                segProfit [curIndex] += order.profit
            #end if
        #end for order
    #end for segment
    return segProfit
# end CalcTotalSegmentProfit    

# Create a new list to store our data and call our OpenFile function to open the file and read it and append the data
orderList = []
fileIn = OpenFile()
# Create for loop to make sure that all the data needed from the file is appended by calling our ReadLine function
for line in fileIn:
    orderList.append(ReadLine(line))
#end for


#call the function to build the orders lists
regLst, segLst, catLst = BuildLists(orderList)

#call the calculation functions by creating new variables to call each one of them
regSalesList = CalcTotalRegionSales(regLst, orderList)
regProfitList = CalcTotalRegionProfit(regLst, orderList)
segSalesList = CalcTotalSegmentSales (segLst, orderList)
segProfitList = CalcTotalSegmentProfit (segLst, orderList)
catSalesList = CalcTotalCategorySales (catLst, orderList)
catProfitList = CalcTotalCategoryProfit (catLst, orderList)

# Create a function to display our menu so our user a can make a selection from our menu 1-7
def MenuChoice(userChoice = None):
        
    # User selects from our menu by inputting an integar 1-7 
    # Defining the menu items. The application relauches the selection menu after an item is selected.
    while userChoice is not None:
        print('\n******************', 'MENU', '******************', )
        print('Please, enter a number 1-7 to select an option from the list below:\n')
        print('1. Region Sales')
        print('2. Region Profit')
        print('3. Segment Sales')
        print('4. Segment Category')
        print('5. Category Sales')
        print('6. Category Category')
        print('7. Exit\n')
        # input("Select An Integar From The Menu (1-7)")
        choice = input('Enter your choice from our menu 1-7: ')

        # Create if-elif-else statements to signify what each of the choices made from our menu (1-7) need to display  
        # and print out to the screen the proper format
        # If 1 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        if choice == '1':
            print('\nRegion Sales')
            for index in range(len(regLst)):
                print(f'{regLst[index]:<20} {locale.currency(regSalesList[index],grouping=True):>20}')
        # If 2 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        elif choice == '2':
            print('\nRegion Profit')
            for index in range(len(regLst)):
                print(f'{regLst[index]:<20} {locale.currency(regProfitList[index],grouping=True):>20}')
        # If 3 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        elif choice == '3':
            print('\nSegment Sales')
            for index in range(len(segLst)):
                print(f'{segLst[index]:<20} {locale.currency(segSalesList[index],grouping=True):>20}')
        # If 4 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        elif choice == '4':
            print('\nSegment Profit')
            for index in range(len(segLst)):
                print(f'{segLst[index]:<20} {locale.currency(segProfitList[index],grouping=True):>20}')
        # If 5 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        elif choice == '5':
            print('\nCategory Sales')
            for index in range(len(catLst)):
                print(f'{catLst[index]:<20} {locale.currency(catSalesList[index],grouping=True):>20}')
        # If 6 is entered display Region Sales descriptions left align 20 character and the values right aligned
        # 20 characters to the right with currecny on them
        elif choice == '6':
            print('\nCategory Profit')
            for index in range(len(catLst)):
                print(f'{catLst[index]:<20} {locale.currency(catProfitList[index],grouping=True):>20}')
        # If 7 is entered break out of the while loop 
        elif choice == '7':
            return print('\nApplication closed')
        
        # In case an invalid menu choice is made the user is prompt to enter an postive number and it has to be 1-7
        else:
            print('\nPlease, enter a valid menu choice 1-7\n')

MenuChoice(0)
```
