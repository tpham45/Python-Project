```python
import csv # Used to write data to a file in csv format

import locale #import locale 
locale.setlocale(locale.LC_ALL,"")

# 1. Let's define open and read data function used to read the input file
def OpenAndReadFile():
    dataRows = open("StoreOrders.csv") #open file

    #skip the first row in the file containing the header        
    next(dataRows) 

    #retain the rest of rows in the file
    dataRows = dataRows.readlines()
  
    return dataRows
  
dataRows = OpenAndReadFile()

#Let's define a helper function to process each row of data and return the colums needed for our calculation
def ProcessDataRow(row):
    
    #Ierate through the each row, take out blank space with strip() method and split each column to form a list of columns
    columns = row.strip("\n").split(",") 
    #Return each column containing the data we need to process
    #columns[-2] corresponds to shippingCost on the spreadsheet
    #columns[-3] corresponds to profit
    #columns[-6] corresponds to sales
    #N.B: The last column correspods to index -1
    return columns[-2], columns[-3], columns[-6]


def OpenAndWriteToFile(dataRows):
  #Here, we define list data structure to hold ALL shippingCosts, profits and sales.
  allShippingCosts, allProfits, allSales = [], [], []
  #Next, we iterate through each data and update the lists above
  for eachLine in dataRows:
    #call the process data function and upack the collumns into shippingCost, profits, and sales
    shippingCost, profit, sales = ProcessDataRow(eachLine)
    allShippingCosts.append(shippingCost)
    allProfits.append(profit)
    allSales.append(sales)
    
  #1. Let's initialize variables to calculate total number of lines, total profit, total sales, and shippingCosts
    totalNumberOfLines = 0
    totalSales = 0 
    totalProfit = 0
    totalShippingCost = 0

  #We assume shippingCosts, profits and sales are all of same length. We therefore iterate through any of them update the above variables

  for index in range(len(allSales)):
    totalNumberOfLines += 1
    totalSales += float(allSales[index])
    totalProfit += float(allProfits[index])
    totalShippingCost += float(allShippingCosts[index])

  #Next, we want to write the report to a csv file
    #Create header 
    header = [
      "Total Number of Lines in the file", 
      "Total Sales",
      "Total Profit",
      "Total Shipping Cost"
  ]
  
  #Create a list to hold all the calculated data
    calculatedData = []
    calculatedData.append([totalNumberOfLines, totalSales, totalProfit, totalShippingCost])

  #Lastly, open and write the data to a csv file called summaryReport
    outputFile = open("summaryReport.csv", 'w', newline='')
    writer = csv.writer(outputFile)

  #write the header into the file
    writer.writerow(header)

  #writer the calcualted data into the file
    writer.writerows(calculatedData)
    outputFile.close()

  #Return the following data to be used in the next function
  return header, calculatedData

header, calculatedData = OpenAndWriteToFile(dataRows)


#Diplay output to the screen with currency being formatted  
def DisplayFormattedOutput(header, calculatedData):
  #Diplays output to the screen with currency being formatted
  for row in calculatedData:
    print ("{:<10} {:<10}\n".format(header[0], row[0]))
    print ("{:<25} {:<30}\n".format(header[1], locale.currency(row[1],grouping = True)))
    print ("{:<25}  {:<30}\n".format(header[2], locale.currency(row[2],grouping = True)))
    print ("{:<25}  {:<30}\n".format(header[3], locale.currency(row[3],grouping = True)))
    

        
DisplayFormattedOutput(header, calculatedData)
```
